```python
import nglview as nv
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
traj = md.load('waterbox_eq.h5')
```

```python
[i.element.symbol for i in 
 traj.topology.atoms if i.residue.index==1]
#getting the elements of the first molecule
```

```python
Os = [i.index for i in traj.topology.atoms if i.element.symbol == 'O']
Hs = [i.index for i in traj.topology.atoms if i.element.symbol == 'H']

#getting the indicies of all the O's and H's in the system
```

```python
grOO = md.compute_rdf(traj[-2000:], 
                          traj.topology.select_pairs(Os, Os))
grHH = md.compute_rdf(traj[-2000:], 
                      traj.topology.select_pairs(Hs, Hs), r_range=[0.165, 1]) #avoiding HH in same molecule
grOH = md.compute_rdf(traj[-2000:], 
                      traj.topology.select_pairs(Os, Hs), r_range=[0.11, 1]) #avoiding OH in same molecule

#calculating OO, HH, and OH RDFs
```

```python
#plotting RDFs

plt.figure(figsize=(16,9))

plt.plot(grOO[0]*10, grOO[1], 'r-', label='OO')
plt.plot(grHH[0]*10, grHH[1], 'g-', label='HH')
plt.plot(grOH[0]*10, grOH[1], 'b-', label='OH')

plt.ylabel('g(r)', fontsize=25)
plt.yticks(fontsize=20)

plt.xlabel('r ($\AA$)', fontsize=20)
plt.xticks(fontsize=20)
plt.xlim(1, 8)

plt.legend(fontsize=20)
```
