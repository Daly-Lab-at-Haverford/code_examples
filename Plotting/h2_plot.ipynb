```python
def fin_diff_2(e, h):
    """
    Returns numerical 2nd derivative at minimum of 
    energy array e
    """
    
    center_loc = np.argmin(e)

    fph = e[center_loc+1]
    fmh = e[center_loc-1]
    f = e[center_loc]
    
    return (fph-2*f+fmh)/h**2

def fin_diff_3(e, h):
    """
    Returns numerical 3rd derivative at minimum of 
    energy array e
    """
    
    center_loc = np.argmin(e)

    num = (-1.0/2.0*e[center_loc-2] + 
           1*e[center_loc-1] + 
           0*e[center_loc] + 
           -1.0*e[center_loc+1] + 
           1.0/2.0*e[center_loc+2])
    
    return num/h**3

def fin_diff_4(e, h):
    """
    Returns numerical 4th derivative at minimum of 
    energy array e
    """
    
    center_loc = np.argmin(e)

    num = (1.0*e[center_loc-2] + 
           -4.0*e[center_loc-1] + 
           6.0*e[center_loc] + 
           -4.0*e[center_loc+1] + 
           1.0*e[center_loc+2])
    
    return num/h**4
```

```python
def harm_pot(k, xmin, x):
    return (1.0/2.0)*k*(x-xmin)**2

def third_pot(k, g, xmin, x):
    return harm_pot(k, xmin, x) + (1.0/6.0)*g*(x-xmin)**3

def fourth_pot(k, g, l, xmin, x):
    return third_pot(k, g, xmin, x) + (1.0/24.0)*l*(x-xmin)**4
```

```python
xmin = 0.742980
x = np.linspace(0.3, 2, 100)

plt.figure(figsize=(16, 9))

k = fin_diff_2(2625.5*(v-v.min()), x[1]-x[0])
plt.plot(x, harm_pot(k, xmin, x), 'g', label='Harmonic Potential', lw=5)

plt.plot(x, 2625.5*(v-v.min()), 'co', label='CCSD(T)/aug-cc-pVTZ')

plt.xlabel(r'$R (\AA)$', fontsize=30)
plt.xticks(fontsize=20)
plt.xlim(0.3, 2)
plt.ylabel(r'$V(R)$ (kJ/mol)', fontsize=30)
plt.yticks(fontsize=20)
plt.ylim(-10, 1250)
plt.legend(fontsize=20)
plt.savefig('h2_harm.pdf')
```

```python
import numpy as np
from scipy import linalg

def T(i, j, dx, redMass):
    a = ((-1.0)**(i-j)/(2.0*redMass*dx**2))
    if i == j:
        return a*np.pi**2/3.0
    else:
        return a*2.0/(i-j)**2

def d(i, j):
    if i == j:
        return 1.0
    else:
        return 0.0

def p(i, j, dx):
    b = ((-1.0)**(i-j))/(dx)
    if i == j:
        return b*0
    else:
        return b*1/(i-j)
    
def dvr1D(dx, energies, redMass):
    """
    Calculate vibrational
    quantum eigenvalues and eigenvectors
    (wavefuctions and energy levels)
    given a bonding potential energy surface.
    This is performed by following Eqn 2.1 of
    https://doi.org/10.1063/1.462100

    Parameters
    ----------
    distances : numpy array
        Sampled bond lengths of potential energy surface.
        The points should be equidistant, ordered,
        and in units of angstroms.

    energies : numpy array
        Sampled energies of potential energy surface.
        The points should be in Hartree.

    redMass : float
        Reduced mass of the bond or vibrational motion
        of interest in atomic mass units.

    Returns
    -------
    eigenvalues : numpy array
        ordered energy levels of potential energy surface.

    eigenvectors : numpy array
        ordered matrix of wavefunctions
    """

    #conversion units
    hart2waveNumber = 2.195*10**5
    amu2eMass = 1822.888
    ang2bohr = 1.88973

    #conversions to hartree atomic units
    redMass = redMass*amu2eMass
    dx = dx*ang2bohr
    nGrid = int(len(energies))

    #fill DVR matrix
    hamGrid = np.zeros((nGrid, nGrid))

    for i in np.arange(nGrid):
        for j in np.arange(nGrid):
            hamGrid[i, j] = T(i, j, dx, redMass) + d(i,j)*energies[i]

    #find eigenvalues, eigenvectors
    els, wfs = linalg.eigh(hamGrid)
    els *= hart2waveNumber
    wfs = wfs.T
    #return output
    return els, wfs
```

```python
import scipy.stats as sps
import scipy.optimize as spo

x = np.linspace(0.3, 2, 100)

def m(x, de, a):
    return de*(1.0-np.exp(-a*(x-0.742980)))**2

popt, pcov = spo.curve_fit(m, x, 2625.5*(v-v.min()), p0=[400, 3], bounds=(0, np.inf))

print(popt)

perr = np.sqrt(np.diag(pcov))

print(perr)

print(sps.pearsonr(2625.5*(v-v.min()), m(x, *popt)))
```
