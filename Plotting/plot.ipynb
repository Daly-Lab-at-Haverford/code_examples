# Making Publication and Thesis Quality Plots

In this notebook we will use the example of calculating then plotting the potential energy surface of H$_2$ to show how one can make and save publication quality plots of data.

First, we import some packages. Matplotlib is one of the most popular libraries for plotting in python. We will use [psi4](https://psicode.org/) to perform some simple energy calculations.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import psi4

#put the calculations in a output file
psi4.set_output_file('output.dat', False)
```

# Potential Energy Surface

H$_2$ is the simplest molecule, made from only two protons and two electrons. Under the Born-Oppenhimer approximation that the electrons adjust instantly to any nuclear motion, the molecule has only one degree of freedom which can change the potential energy. This degree of freedom is the distance between the two protons, which is called the bond length, $R$. 

Here, we will vary the bond legnth and use psi4 to calculate the energy. This calculation should take just long enough that it would be really annoying to have to run it again.

```python
R = np.linspace(0.3, 2, 100) # Generate 100 bond lengths between 0.3 and 2 Angstroms
V = np.zeros(100) # An array of zeros. I plan to put the potential energies here.

# psi4 takes the molecular geometry as a string. I've defined this below as the variable h2.
# the geometry is formatted as xyz coordinates
h2 = """
0 1
H 0 0 0
H 0 0 {}
"""
# first line: charge, multiplicity.
# second line: atom symbol, XYZ coordinates for the first H
# third line: atom symbol, XYZ coordinates for the second H
# h2.format('x') will replace the {} with x,
# so I can put whatever number I want in for the Z coordinate of atom 2

for bond_length_index, bond_length in enumerate(R):
    # for ... enumerate(R) will produce an index and a value from R for every value in R
    # so, I will get an index and bond length for every value in R
    h2geo = psi4.geometry(h2.format(bond_length))
    # make a psi4 geometry object with the formatted string
    V[bond_length_index] = psi4.energy('ccsd(t)/aug-cc-pVTZ')
    # perform a psi4 energy calculation. By default, it will perform this calculation on the most recent
    # structure that psi4 has seen.
```

# Harmonic Approximation

Most bonds spend most of their time near the energy minimum. Near this minimum, the bond appears approximately harmonic,

$$ V(R) = \frac{1}{2}k(R-R_0)^2 $$

where $R_0$ is the bond legnth where the energy is minimized and $k$ is a constant describing the bond stregnth. It can be shown that the best possible harmonic approximation is obtained if $k$ is given by,

$$ k = \left(\frac{\partial^2 V(R)}{\partial R^2}\right)_{R=R_0} $$

Let's set up some functions to get this from our data.

```python
def fin_diff_2(e, h):
    """
    Returns numerical 2nd derivative at minimum of 
    energy array e by the central finite difference method
    more info: https://en.wikipedia.org/wiki/Finite_difference#:~:text=Second%2Dorder%20central

    Parameters
    ----------
    e: numpy array
        Sampled potential energy surface.

    energies : float
        distance between sampled bond lengths

    Returns
    -------
    finite difference second derivative centered at minimum energy point
    """

    #find the location where the energy is minimized
    center_loc = np.argmin(e)

    #find the energy of the point to the right of the minimum
    fph = e[center_loc+1]
    #find the energy of the point to the right of the minimum
    fmh = e[center_loc-1]
    #find the energy of the minimum
    f = e[center_loc]

    #calculate and return the second derivative
    return (fph-2*f+fmh)/h**2

def harm_pot(k, x_0, x):
    """
    Returns energy according to harmonic potential

    Parameters
    ----------
    k: float
        spring constant. Should have units of energy/distance**2

    x_0: float
        equlibrium or minimum energy distance
        should have units of distance

    x: float or ndarray
        distance(s) where the energy should be calculated.
        should have units of distance.

    Returns
    -------
    harmonic energy/energies for the distance(s) 
    """ 
    # just calculating it in one line.
    return (1.0/2.0)*k*(x-x_0)**2
```

# Plotting - the important part

Now that we have some data, let's plot it! Here are some guidelines for making publication quality plots. There are specific cases where every one of these rules can be thrown out, but those moments are rare.

* The default size for a plot should be 16 inches wide by 9 inches high. 
    * In some cases, it will make more sense to have a square plot. The most common case is when the x-values and the y-values have some direct comparative relationship.
* Use a large font for axes titles and tick labels. I reccommend 30 for axes titles and 25 for tick labels.
* Make your lines thick (linewidth of 2 or 3) and your markers big (markersize around 15).
* Do NOT use a color for the background of the plot.
* Avoid using a legend unless necessary.
* Do NOT use gridlines.
* **LABEL ALL AXES WITH UNITS!**
* Do NOT connect lines between points unless there is a good scientific or communicative reason. Examples of good reasons:
    * The connecting line(s) represent a real mathematical fit to the data. 
    * The line itself has a physical meaning.
    * The points are very closely spaced, like in a spectrum.
    * The data would be difficult to follow without the line. In this case, use a dotted or dashed line so that it is clear that there is no physical meaning to the line and that it is mearly communicative.
* Avoid giving the plot a title. The plot will appear with a caption in its final location.
* Save your plot as a PDF or in another vector graphics format.
    * If this is somehow not possible, save the figure with a resolution of at least 350 DPI.
* If you're using the same labels, colors, etc in multiple plots, put the labels in a list before

```python
#defining variables and other random things I might need later

#I happen to know from prior calculations that this is the value of R_0 for this QC method
R_0 = 0.742980
R = np.linspace(0.3, 2, 100) # Generate 100 bond lengths between 0.3 and 2 Angstroms
# Even though I already defined R, I'm redefining it in this cell so it's easy to make changes if I need to

#plot the PES data
plt.plot(R, (V-V.min()))
```

```python
# set the figure size. 
plt.figure(figsize=(16, 9))

#defining variables and other random things I might need later

#I happen to know from prior calculations that this is the value of R_0 for this QC method
R_0 = 0.742980
R = np.linspace(0.3, 2, 100) # Generate 100 bond lengths between 0.3 and 2 Angstroms
# Even though I already defined R, I'm redefining it in this cell so it's easy to make changes if I need to

#calculate k using fin_diff function
k = fin_diff_2(V, np.diff(R)[0])

#plot the PES data. ms means "marker size"
plt.plot(R, (V-V.min()), 'bo', label='CCSD(T)/aug-cc-pVTZ', ms=9)
#plot the harmonic potential
plt.plot(R, harm_pot(k, R_0, R), 'k', label='Harmonic Potential', lw=3)

#set the xlabel. Notice that I can use latex to get needed math text.
plt.xlabel(r'$R$ ($\AA$)', fontsize=30)
#increasing the font size for the xticks
plt.xticks(fontsize=25)
#adjusting the xlimits
plt.xlim(0.3, 2)
#set the ylabel, again using latex for math
plt.ylabel(r'$V(R)$ ($E_h$)', fontsize=30)
#increasing the font size for the yticks
plt.yticks(fontsize=25)
#adjusting the ylimits
plt.ylim(0, 0.5)
#add a legend with a solid fontsize
plt.legend(fontsize=25)
#save the figure as a pdf
plt.savefig('h2_harm.pdf', transparent=True)
```

# Saving Plot Data

Great, we have a publication quality plot! However, what if you closed this notebook and later needed to change the y-axis units to kJ/mol? Or you just wanted to see if a red line might be better than a black line? Well, you'd need to perform all the calculations all over again. This is very inefficient. 

Instead, we can save the data that goes into the plot, and write a separate script which allows us to make small edits to the plot quickly. First, we need to save the data.

As a reminder, the code we used to plot the data were:

`plt.plot(R, (V-V.min()), 'bo', label='CCSD(T)/aug-cc-pVTZ', ms=9)`

and

`plt.plot(R, harm_pot(k, R_0, R), 'k', label='Harmonic Potential', lw=3)`

Fortunately, all our data was in numpy arrays, which makes saving it very easy. We just need to use np.save().

```python
#save R data in a file called 'bond_lengths.npy'
np.save('bond_lengths', R)
#save PES data in a file called 'pes.npy'
np.save('pes', V)
#save harmonic potential in a file called 'hp.npy'
np.save('hp', harm_pot(k, R_0, R))

#we only need 3 save statements because the plots share the same x-axis values
```

# Loading Plot Data

Let's imagine that a week later, we decided that we wanted to change the colors to black and red to match the Haverford colors. Now, we don't need to recalculate everything. Let's just use `np.load()` to import the data we saved before.

```python
#reimport stuff
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
#I don't need psi4, because I'm not doing any calculations.
```

```python
R = np.load('bond_lengths.npy')
V = np.load('pes.npy')
#notice I'm making a new variable for the harmonic potential data. 
#this will change how I plot the data later
hp = np.load('hp.npy')
```

```python
# set the figure size. 
plt.figure(figsize=(16, 9))

#plot the PES data. ms means "marker size"
plt.plot(R, (V-V.min()), 'ro', label='CCSD(T)/aug-cc-pVTZ', ms=9)
#plot the harmonic potential. notice this is different from before
plt.plot(R, hp, 'k', label='Harmonic Potential', lw=3)

#set the xlabel. Notice that I can use latex to get needed math text.
plt.xlabel(r'$R$ ($\AA$)', fontsize=30)
#increasing the font size for the xticks
plt.xticks(fontsize=25)
#adjusting the xlimits
plt.xlim(0.3, 2)
#set the ylabel, again using latex for math
plt.ylabel(r'$V(R)$ ($E_h$)', fontsize=30)
#increasing the font size for the yticks
plt.yticks(fontsize=25)
#adjusting the ylimits
plt.ylim(0, 0.5)
#add a legend with a solid fontsize
plt.legend(fontsize=25)
#save the figure as a pdf
plt.savefig('h2_harm.pdf', transparent=True)
```
